```python
def get_text_link_from_sel(sel):
    mylist = []
    try:
        results = r.html.find(sel)
        for result in results:
            mytext = result.text
            mylink = list(result.absolute_links)[0]
            mylist.append((mytext, mylink))
        return mylist
    except:
        return None
```

```python
def get_url_company(url):
    return 'http://query.sse.com.cn/commonQuery.do?jsonCallBack=jsonpCallback00001&isPagination=false&sqlId=COMMON_SSE_ZQPZ_GP_GPLB_C&productid='+url+'&_=11111111111111'

def get_url_executive(url):
    return 'http://query.sse.com.cn/commonQuery.do?jsonCallBack=jsonpCallback80740&isPagination=true&productid='+url+'&sqlId=COMMON_SSE_ZQPZ_GG_GGRYLB_L&pageHelp.pageSize=25&pageHelp.pageCount=50&pageHelp.pageNo=1&pageHelp.beginPage=1&pageHelp.cacheSize=1&pageHelp.endPage=5&_=1535506282048'

def collect_sse():#调取30页，相当于主函数
    company_code=[]
    company_abbr=[]
    fullname=[]
    executives=[]
    business=[]
    start_time=[]
    index=1
    for x in company_codes:
        url_company=get_url_company(x)
        response_company=requests.get(url_company,headers=header_company)
        response_company=response_company.text[19:-1]
        json_company=json.loads(response_company)
        COMPANY_CODE = jsonpath(json_company,'$..result..COMPANY_CODE')
        COMPANY_ABBR = jsonpath(json_company,'$..result..COMPANY_ABBR')
        FULLNAME = jsonpath(json_company,'$..result..FULLNAME')
        
        url_executives=get_url_executive(x)
        response_executives=requests.get(url_executives,headers=header_company)
        response_executives=response_executives.text[19:-1]
        json_executives=json.loads(response_executives)
        EXECUTIVES = jsonpath(json_executives,'$..result..NAME')
        BUSINESS = jsonpath(json_executives,'$..result..BUSINESS')
        START_TIME = jsonpath(json_executives,'$..result..START_TIME')
        
        company_code.append(COMPANY_CODE)
        company_abbr.append(COMPANY_ABBR)
        fullname.append(FULLNAME)
        executives.append(EXECUTIVES)
        business.append(BUSINESS)
        start_time.append(START_TIME)
        print (index, COMPANY_CODE,COMPANY_ABBR,FULLNAME,EXECUTIVES,BUSINESS,START_TIME)
        index+=1
        
    return (company_code,company_abbr,fullname,executives,business,start_time)
```
